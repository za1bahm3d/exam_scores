# Clean and transform

Create numpy arrays

```python
import pandas as pd
import numpy as np

url = "https://raw.githubusercontent.com/lilaceri/Working-with-data-/main/Data%20Sets%20for%20code%20divisio/exams.csv"
df = pd.read_csv(url)
df.info()

# Create NumPy arrays
math_scores = df['math_score'].to_numpy()
reading_scores = df['reading_score'].to_numpy()
writing_scores = df['writing_score'].to_numpy()
```

calculate range, mean, standard deviation, median, interquartile range Create a new numpy array to hold the average exam score (of the three scores for each student)

```python
import numpy as np

# Calculate average exam score for each student
average_scores = (math_scores + reading_scores + writing_scores) / 3

# Function to calculate required statistics
def calculate_stats(scores):
    stats = {
        'range': np.ptp(scores),
        'mean': np.mean(scores),
        'standard_deviation': np.std(scores),
        'median': np.median(scores),
        'interquartile_range': np.percentile(scores, 75) - np.percentile(scores, 25)
    }
    return stats

# Calculating statistics for each subject
math_stats = calculate_stats(math_scores)
reading_stats = calculate_stats(reading_scores)
writing_stats = calculate_stats(writing_scores)
average_stats = calculate_stats(average_scores)

# Displaying the results
print("Math Statistics:", math_stats)
print("Reading Statistics:", reading_stats)
print("Writing Statistics:", writing_stats)
print("Average Exam Score Statistics:", average_stats)
```

Round to 2 decimal places

```python
# Function with rounding
def calculate_stats(scores):
    stats = {
        'range': round(np.ptp(scores), 2),
        'mean': round(np.mean(scores), 2),
        'standard_deviation': round(np.std(scores), 2),
        'median': round(np.median(scores), 2),
        'interquartile_range': round(np.percentile(scores, 75) - np.percentile(scores, 25), 2)
    }
    return stats

# Recalculate the stats with rounding
math_stats = calculate_stats(math_scores)
reading_stats = calculate_stats(reading_scores)
writing_stats = calculate_stats(writing_scores)
average_stats = calculate_stats(average_scores)

# Display the rounded results
print("Math Statistics:", math_stats)
print("Reading Statistics:", reading_stats)
print("Writing Statistics:", writing_stats)
print("Average Exam Score Statistics:", average_stats)
```

Using a loop, convert each row into a dictionary and add it to a new list so that you have a list of dictionaries

```python
# Using pandas built-in method to convert the whole DataFrame to a list of dictionaries
records_list = df.to_dict(orient='records')

# Display the first few records
print(records_list[:3])
```

Create two new lists, one containing all the records for males and one for females

```python
# Separate male and female records
male_records = [record for record in records_list if record['gender'] == 'male']
female_records = [record for record in records_list if record['gender'] == 'female']

# Display the first few records from each list to confirm
print("First 3 Male Records:", male_records[:3])
print("First 3 Female Records:", female_records[:3])
```

Calculate mean scores for each exam for males, and mean scores for each exam for females

```python
import numpy as np

# Extract scores for males
male_math_scores = [record['math_score'] for record in male_records]
male_reading_scores = [record['reading_score'] for record in male_records]
male_writing_scores = [record['writing_score'] for record in male_records]

# Extract scores for females
female_math_scores = [record['math_score'] for record in female_records]
female_reading_scores = [record['reading_score'] for record in female_records]
female_writing_scores = [record['writing_score'] for record in female_records]

# Calculate mean scores
male_means = {
    'math': round(np.mean(male_math_scores), 2),
    'reading': round(np.mean(male_reading_scores), 2),
    'writing': round(np.mean(male_writing_scores), 2)
}

female_means = {
    'math': round(np.mean(female_math_scores), 2),
    'reading': round(np.mean(female_reading_scores), 2),
    'writing': round(np.mean(female_writing_scores), 2)
}

# Display the mean scores
print("Male Mean Scores:", male_means)
print("Female Mean Scores:", female_means)
```

Compare each average with the overall averages

```python
# Overall average scores from the original dataframe
overall_math_mean = round(df['math_score'].mean(), 2)
overall_reading_mean = round(df['reading_score'].mean(), 2)
overall_writing_mean = round(df['writing_score'].mean(), 2)

overall_means = {
    'math': overall_math_mean,
    'reading': overall_reading_mean,
    'writing': overall_writing_mean
}

print("Overall Mean Scores:", overall_means)
```

```python
# Comparing male and female averages with overall averages
comparison = {
    'math': {
        'male_diff': round(male_means['math'] - overall_means['math'], 2),
        'female_diff': round(female_means['math'] - overall_means['math'], 2)
    },
    'reading': {
        'male_diff': round(male_means['reading'] - overall_means['reading'], 2),
        'female_diff': round(female_means['reading'] - overall_means['reading'], 2)
    },
    'writing': {
        'male_diff': round(male_means['writing'] - overall_means['writing'], 2),
        'female_diff': round(female_means['writing'] - overall_means['writing'], 2)
    }
}

# Display the comparison
print("Comparison with Overall Averages:", comparison)
```

Compare with ethnicity

```python
# Group the data by ethnicity and calculate the mean scores
ethnicity_means = df.groupby('ethnicity')[['math_score', 'reading_score', 'writing_score']].mean().round(2)

# Display mean scores by ethnicity
print("Mean Scores by Ethnicity:\n", ethnicity_means)
```

```python
# Initialize a dictionary to store comparisons
ethnicity_comparison = {}

# Loop through each ethnicity and compare with overall means
for ethnicity in ethnicity_means.index:
    ethnicity_comparison[ethnicity] = {
        'math_diff': round(ethnicity_means.loc[ethnicity, 'math_score'] - overall_means['math'], 2),
        'reading_diff': round(ethnicity_means.loc[ethnicity, 'reading_score'] - overall_means['reading'], 2),
        'writing_diff': round(ethnicity_means.loc[ethnicity, 'writing_score'] - overall_means['writing'], 2)
    }

# Display the comparison
print("\nComparison with Overall Averages by Ethnicity:\n", ethnicity_comparison)
```

Histogram for math scores and reading scores to see their distributions

```python
import matplotlib.pyplot as plt

# Plot histograms
plt.figure(figsize=(12, 5))

# Math Score Distribution
plt.subplot(1, 2, 1)
plt.hist(df['math_score'], bins=10, edgecolor='black', alpha=0.7)
plt.title('Math Score Distribution')
plt.xlabel('Score')
plt.ylabel('Frequency')

# Reading Score Distribution
plt.subplot(1, 2, 2)
plt.hist(df['reading_score'], bins=10, edgecolor='black', alpha=0.7)
plt.title('Reading Score Distribution')
plt.xlabel('Score')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

Scatter plots to analyze the relationship between math scores vs reading scores, reading vs writing scores and maths vs writing scores.

```python
# Import necessary libraries
import matplotlib.pyplot as plt

# Function to plot scatter with regression line using Matplotlib only
def plot_scatter(x, y, x_label, y_label, title, color):
    plt.figure(figsize=(7, 5))

    # Scatter plot
    plt.scatter(df[x], df[y], alpha=0.5, color=color, label="Data Points")

    # Calculate and plot regression line
    m, b = np.polyfit(df[x], df[y], 1)  # Linear regression
    plt.plot(df[x], m * df[x] + b, color="red", label="Regression Line")

    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.legend()
    plt.grid(True)
    plt.show()

# Scatter plot for Reading vs Writing Scores
plot_scatter('reading_score', 'writing_score', "Reading Score", "Writing Score", "Reading vs Writing Scores (With Regression Line)", "orange")

# Scatter plot for Math vs Writing Scores
plot_scatter('math_score', 'writing_score', "Math Score", "Writing Score", "Math vs Writing Scores (With Regression Line)", "green")

# Scatter plot for Math vs Reading Scores
plot_scatter('math_score', 'reading_score', "Math Score", "Reading Score", "Math vs Reading Scores (With Regression Line)", "blue")


```

We'll check how strongly math, reading, and writing scores correlate:

```python
# Calculate correlation matrix
correlation_matrix = df[['math_score', 'reading_score', 'writing_score']].corr().round(2)
print("\nCorrelation Matrix:\n", correlation_matrix)
```

We will compare the correlation of scores for:

Students who receive free school meals (lunch == 'free/reduced')
Students who took a preparation course (preparation_course == 'completed')

```python
# Correlation for students with free school meals
free_meals_df = df[df['lunch'] == 'free/reduced']
free_meals_corr = free_meals_df[['math_score', 'reading_score', 'writing_score']].corr().round(2)
print("\nCorrelation Matrix for Free School Meals Students:\n", free_meals_corr)

# Correlation for students who took a preparation course
prep_course_df = df[df['preparation_course'] == 'completed']
prep_course_corr = prep_course_df[['math_score', 'reading_score', 'writing_score']].corr().round(2)
print("\nCorrelation Matrix for Students Who Completed Preparation Course:\n", prep_course_corr)
```

School Meal Students vs Exam Scores (Boxplots)

```python
# Import necessary libraries
import matplotlib.pyplot as plt

# Filter students based on lunch type (Free/Reduced vs. Standard)
free_meal_students = df[df['lunch'] == 'free/reduced']
standard_meal_students = df[df['lunch'] == 'standard']

# Function to create boxplots for comparison
def plot_boxplot(data1, data2, labels, title, y_label, color1, color2):
    plt.figure(figsize=(7, 5))
    plt.boxplot([data1, data2], labels=labels, patch_artist=True,
                boxprops=dict(facecolor=color1, alpha=0.6),
                medianprops=dict(color="black"))

    plt.title(title)
    plt.ylabel(y_label)
    plt.grid(True)
    plt.show()

# Boxplot for Math Scores
plot_boxplot(free_meal_students['math_score'], standard_meal_students['math_score'],
             ['Free Meal', 'Standard Meal'], "Math Scores Comparison", "Math Score", "red", "blue")

# Boxplot for Reading Scores
plot_boxplot(free_meal_students['reading_score'], standard_meal_students['reading_score'],
             ['Free Meal', 'Standard Meal'], "Reading Scores Comparison", "Reading Score", "red", "blue")

# Boxplot for Writing Scores
plot_boxplot(free_meal_students['writing_score'], standard_meal_students['writing_score'],
             ['Free Meal', 'Standard Meal'], "Writing Scores Comparison", "Writing Score", "red", "blue")
```

Exam Scores vs Preparation Course


```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Function to create boxplots for comparison of preparation course vs exam scores
def plot_prep_course_boxplot(df, score_column, title, y_label, color1, color2):
    # Filter students based on whether they completed the preparation course
    prep_completed = df[df['preparation_course'] == 'completed'][score_column]
    prep_none = df[df['preparation_course'] == 'none'][score_column]

    # Create a boxplot
    plt.figure(figsize=(7, 5))
    plt.boxplot([prep_completed, prep_none], labels=['Completed', 'Not Completed'], patch_artist=True,
                boxprops=dict(facecolor=color1, alpha=0.6),
                medianprops=dict(color="black"))

    plt.title(title)
    plt.ylabel(y_label)
    plt.grid(True)
    plt.show()

# Boxplot for Math Scores vs Preparation Course
plot_prep_course_boxplot(df, 'math_score', "Math Scores vs Preparation Course", "Math Score", "green", "red")

# Boxplot for Reading Scores vs Preparation Course
plot_prep_course_boxplot(df, 'reading_score', "Reading Scores vs Preparation Course", "Reading Score", "blue", "red")

# Boxplot for Writing Scores vs Preparation Course
plot_prep_course_boxplot(df, 'writing_score', "Writing Scores vs Preparation Course", "Writing Score", "purple", "red")
```

**Summary of Findings:**

Students on free school meals tend to score lower in math, reading, and writing, with greater score variability and more low outliers. This suggests that economic background may impact academic performance.

Students who completed a preparation course scored higher across all subjects, with higher median scores and fewer low outliers. This indicates that prep courses are effective in improving student performance.

Reading and Writing scores have the strongest correlation (0.96), meaning students who excel in one tend to perform well in the other. Math also shows a strong correlation with reading and writing (0.81), suggesting that strong literacy skills may contribute to better math performance.

Most students score between 60-80, as shown in histograms, but some struggle significantly or perform exceptionally well, highlighting learning disparities.

Key Takeaways: Socio-economic background impacts scores, preparation courses improve performance, and strong literacy skills are linked to success in math. Schools can provide extra support for free meal students, expand access to prep courses, and emphasize literacy development to improve academic outcomes.
